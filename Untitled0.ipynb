```python
import pandas as pd

df = pd.read_csv("marketing_campaign.csv", sep="\t")
print("Original shape:",df.shape)
df.head()
```

```python
print("Before removing duplicates:", df.shape)
df = df.drop_duplicates()
print("After removing duplicates:", df.shape)
```

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y')
print(df['Dt_Customer'].head())
```

```python
print("Missing values before filling:")
print(df.isnull().sum())

# Fill missing 'Income' with mean
df['Income'] = df['Income'].fillna(df['Income'].mean())



# Verify
print("\nMissing values after cleaning:")
print(df.isnull().sum())
```

```python
print("Original column names:\n", df.columns.tolist())
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
print("Cleaned column names:\n", df.columns.tolist())
```

```python
print("Data types before:\n", df.dtypes)
df['income'] = df['income'].astype(int)
print("Data types after:\n", df.dtypes)
```

```python
df.to_csv("marketing_campaign_cleaned.csv", index=False)
print("✅ Cleaned dataset saved as 'marketing_campaign_cleaned.csv'")
```
